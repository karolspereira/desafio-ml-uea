```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
retorno = load_iris()
```

```python
retorno
```

```python
def defineNames(target):
    if(target == 0):
       return 'setosa'
    elif(target == 1):
        return 'versicolor'
    elif(target == 2):
         return 'virginica'
```

```python
iris = pd.DataFrame(retorno.data)
iris.columns = ['comp sepal','larg sepal','comp petal','larg petal']
```

```python
iris['target'] = retorno.target
iris['target_name'] = iris['target'].apply(defineNames)
```

```python
display(iris)
```

```python
iris.describe()
```

```python
#plotar boxplot

import seaborn as sns
sns.set(rc={'figure.figsize':(15,12)})
sns.set_theme(style="white")
```

```python
ax= sns.boxplot(data=iris)
```

```python
#Verificar a presença de dados nulos
iris.info()
```

```python
#verificar outliers
iris[(iris['larg sepal'] <2.2) | (iris['larg sepal'] > 4)]
```

```python
#verificar a correlação entre variáveis
#corr=1 (correlação forte) corr=-1 (crescimento inverso) corr=0 (correlação fraca)
iris.corr()
```

```python
ax = sns.pairplot(iris)
```

```python
ax = sns.pairplot(iris, hue ='target_name')
```

```python
sns.scatterplot(x='larg petal', y='comp petal',data=iris, hue='target_name', palette=['green','orange','blue'])
plt.show()
                
```

```python
#Separando em treino e teste
```

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(retorno.data, retorno.target, random_state=1)
```

```python
treino = pd.DataFrame(x_treino)
treino.columns = ['comp sepal', 'larg sepal', 'comp petal', 'larg petal']
treino['target'] = y_treino
```

```python
display(treino)
```

```python
#plotar gráfico para dados de treino
sns.scatterplot(x='larg petal',y='comp petal', data=treino, hue='target', palette=['green','orange', 'blue'])
plt.show()
```

```python
#separar os targets
treino[(treino['target']== 2) & ((treino['larg petal'] == 1.7) | (treino['larg petal']==1.5))]
```

```python
#traçando duas retas
sns.scatterplot(x='larg petal',y='comp petal',data=treino, hue='target', palette=['green', 'orange','blue'])
x_verde = [0,1.4]
y_verde = [4.8,0]
sns.lineplot(x=x_verde, y=y_verde,color='g')
x_azul = [0,4.9]
y_azul = [7.7,0]
sns.lineplot(x=x_azul, y=y_azul, color='b')
plt.show()
```

```python
#tratamento de dados de teste
teste = pd.DataFrame(x_teste)
teste.columns = ['comp sepal', 'larg sepal', 'comp petal', 'larg petal']
teste['target'] = y_teste
```

```python
display(teste)
```

```python
sns.scatterplot(x='larg petal',y='comp petal',data=teste, hue='target', palette=['green', 'orange','blue'])
x_verde = [0,1.4]
y_verde = [4.8,0]
sns.lineplot(x=x_verde, y=y_verde,color='g')
x_azul = [0,4.9]
y_azul = [7.7,0]
sns.lineplot(x=x_azul, y=y_azul, color='b')
plt.show()
```

```python
#Usando Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
modelo = RandomForestClassifier()
```

```python
modelo.fit(x_treino,y_treino)
```

```python
predicao = modelo.predict(x_teste)
```

```python
predicao
```

```python
teste['predicao'] = predicao
teste['check'] = teste['target'] == teste['predicao']
display(teste)
```

```python
teste[teste.check == False]
```

```python
from sklearn.metrics import classification_report
print(classification_report(teste['target'], teste['predicao'], target_names=["setosa","versicolor","virginica"]))
```
